### Scraping job sites for information on Data Analysts and Data Engineers

```python
import duckdb
from jobspy import scrape_jobs
```

```python
analyst_jobs = scrape_jobs(
    site_name=["indeed","linkedin","glassdoor"],
    search_term="data analyst",
    location="United Kingdom",
    results_wanted=30, #total number of results are 30x3
    hours_old=72, # (only Linkedin/Indeed is hour specific, others round up to days old)
    country_indeed="UK",  # only needed for indeed / glassdoor
    linkedin_fetch_description=True)
```

```python
de_jobs = scrape_jobs(
    site_name=["indeed","linkedin","glassdoor"],
    search_term="data engineer",
    location="United Kingdom",
    results_wanted=30,
    hours_old=72, # (only Linkedin/Indeed is hour specific, others round up to days old)
    country_indeed="United Kingdom",  # only needed for indeed / glassdoor
    linkedin_fetch_description=True)
```

```python
con = duckdb.connect(database = "jobinfo.db")
con.execute("INSERT INTO jobinfo SELECT * FROM analyst_jobs UNION SELECT * FROM de_jobs")
```

```python
con.sql("SELECT * FROM jobinfo ORDER BY title DESC")
```
